```python
import pandas as pd
```

```python
df= pd.DataFrame([[0, 1], [2, 3]],index=['cat', 'dog'],columns=['weight', 'height'])
```

```python
print('{:*^80}'.format('dataframe'))
df
```

```python
#Then df.stack() turns our single-level column df into a dataseries with a multi-index index by 
#fitting the columns into a new inner index (index level 1) for each value in the 
#old outer index (index level 0). The outer level looks like it has only 2 values [cat,dog], but 
#it is only for neat display and is really 4 values [cat,cat,dog,dog]
print('{:*^80}'.format('stacked dataframe'))
df.stack()
```

```python
print('{:*^80}'.format('stacked dataframe with index in column'))
stacked = df.stack().reset_index(level = 1)
stacked
```

```python
print('{:*^80}'.format('pivot_table recovered original dataframe (with extra name for columns)'))
recovered_df1 = df.stack().reset_index(level = 1).pivot_table(index = stacked.index, columns = 'level_1',values = 0)
recovered_df1
```

```python
# You can pass a header argument into pandas.read_excel() that indicates how many rows are to be used as 
# headers. In your particular case, you'd want header=[0, 1], indicating the first two rows.
#index_col = 0 means we are considering the first column as index
df = pd.read_excel(r'D:\Ankit\Pandas\12_stack\stocks.xlsx',header=[0,1],index_col=0)
```

```python
df
```

```python
#Stack brings the columns to rows
#df.stack(level=0) - 0 brings the first header into columns / 1 would bring the second header into columns
stacked = df.stack()
stacked
```

```python
#to get your original dataframe back
stacked.unstack()
```
