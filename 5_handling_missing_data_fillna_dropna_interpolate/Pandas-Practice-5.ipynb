```python
df.set_index('day',inplace=True)
```

<b>Methods to handle missing values</b>

```python
#filling all missing values with 0
new_df = df.fillna(0)
new_df
```

```python
#filling missing values with appropriate values column wise
new_df2 = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df2
```

```python
#fillna forward fill -->if missing value is detected,it would fill the 
#value above the missing value(carry-forward previous day's value)
new_df3 = df.fillna(method='ffill')
new_df3
```

```python
#similary bfill -> fills the value of the missing element by filling it with the value after it
new_df4 = df.fillna(method='bfill')
new_df4
```

```python
#If you want to replace missing values with the values in their left or right columns, then use axis="columns"
new_df5 = df.fillna(method='ffill',axis="columns")
new_df5
```

```python
new_df6 = df.fillna(method='bfill',axis="columns")
new_df6
```

```python
#Suppose while using ffill or bfil, there are continuous data missing and you want to add limit in the filling 
#of carry-forward and carry backward values for filling the missing data
new_df7 = df.fillna(method='ffill',axis="columns",limit=1)
new_df7
```

INTERPOLATION:
Interpolation is a technique in Python used to estimate unknown data points between two known data points.


```python
df.interpolate(method ='linear')
```

```python
#method='time' considers the fact that if the date of the missing value is closer to the date from which it 
#will be taking value, then it will fill a number in missing value closer to the value it is taking as dates are closer
#Similarly the missing value will be filed with a number farther than the actual value if the difference between dates are more
df.interpolate(method='time')
```

```python
#To drop rows with misisng values
drop_df = df.dropna() #->by default how="any" means the row will get dropped if there is any missing value
drop_df
```

```python

drop_df = df.dropna(how="all") #->how="all" means the row will get dropped if all the row got missing value
drop_df
```

```python
drop_df = df.dropna(thresh=1) #means keep the rows if there is 1 non-missing value else drop
drop_df
```

```python
drop_df = df.dropna(thresh=2) #means keep the rows if there is 1 non-missing value else drop
drop_df
```
