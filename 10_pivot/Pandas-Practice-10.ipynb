```python
df = pd.read_csv(r'D:\Ankit\Pandas\10_pivot\weather2.csv')
```

```python
df
```

```python
#Mean of temparatures of each city in each day #mean is default
df.pivot_table(values='temperature', index='date', columns='city', aggfunc='mean')
```

```python
#margins structures the dataframe and puts "ALL" columns which aggregartes aggfunc row wise
df.pivot_table(index='date', columns='city', aggfunc='sum',margins=True)
```
