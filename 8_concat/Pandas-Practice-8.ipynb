```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temparature":[32,45,30],
    "humidity":[80,60,78]
})
```

```python
df2 = pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temparature":[68,65,75],
    "humidity":[21,14,35]
})
```

```python
#concate both the dataframes (use parameter axis=1 to concat it column-wise)
df_merge = pd.concat([df1,df2],axis=1)
df_merge
```

```python
#concate both the dataframes (default = row-wise) (use parameter axis=1 to concat it column-wise)
df_merge = pd.concat([df1,df2])
```

```python
df_merge
```

```python
#as you can see the index are not continuous as the index are picked from their original tables
#we need to make it continuous
df_merge2 = pd.concat([df1,df2],ignore_index=True)
df_merge2
```

```python
#To highlight each dataframe using keys (doesn't work with ignore_index)
df_merge3 = pd.concat([df1,df2],keys=["india","us"])
df_merge3
```

```python
#to select a particular key in the merged dataframe
df_merge3.loc["india"]
```

```python
#creating two dataframes with a column city column
#We use index so that those rows will be mapped with the index while concatenating
temp_df = pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temparature":[32,45,30],
    "index":[0,1,2]
})
```

```python
temp_df
```

```python
#We use index so that those rows will be mapped with the index while concatenating
windspeed_df = pd.DataFrame({
    "city":["bangalore","delhi"],
    "windspeed":[7,23],
    "index":[2,1]
})
```

```python
windspeed_df
```

```python
#Concat the tables
df_concat = pd.concat([temp_df,windspeed_df],axis=1)
df_concat
```

<b>MERGE </b>

We can use merge to merge two dataframes on a row

```python
df_merges = pd.merge(temp_df,windspeed_df,how='outer',on='city')
df_merges
```
