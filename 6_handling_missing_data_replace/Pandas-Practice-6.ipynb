```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:\Ankit\Pandas\6_handling_missing_data_replace\weather_data.csv')
```

```python
df
```

```python
#The missing values are -99999 and event=0 instead of NaN
#So we need to replace these values with NaN
#Colum specific replacement with NaN
new_df = df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.NaN)
```

```python
new_df
```

```python
#Filling missing values with different values
new_df2 = df.replace({
    -99999:np.NaN,
    '0':'Sunny'
})
```

```python
new_df2
```

```python
#Regex example
df_2 = pd.read_csv(r'D:\Ankit\Pandas\6_handling_missing_data_replace\weather_data(regex).csv')
```

```python
df_2
```

```python
#the above dataframe has F/C in temperature colm and mph in windspeed with we need to remove 
new_df3 = df_2.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
```

```python
new_df3
```
